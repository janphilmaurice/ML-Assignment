```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# For accuracy of different combination of hyper
hyper_acc=[
    0.8517999999999999,
    0.8398,
    0.8538,
    0.8588000000000001,
    0.8593999999999999,
    0.8664,
    0.8506,
    0.8573999999999999,
    0.8539999999999999,
    0.861,
    0.8462,
    0.8665999999999998,
    0.706,
    0.857,
    0.7918000000000001,
    0.8573999999999999,
    0.7458,
    0.8596
]

print(hyper_acc.index(max(hyper_acc)))

fig, axs = plt.subplots(1, 1, figsize=(8, 6))
labels=range(0,len(hyper_acc),1)

acc=[round(x, 2) for x in hyper_acc]
colors = ['blue'] * (len(hyper_acc)-1)
colors[11] = 'red' 

# 子图1
bars = axs.bar(labels, acc, alpha=0.6)
axs.plot(labels, acc, label="Mean Reliability", marker="s")
axs.set_title("Combination of hyperparameter",fontsize=14, fontweight='bold')
axs.spines['top'].set_linewidth(2)
axs.spines['right'].set_linewidth(2)
axs.spines['bottom'].set_linewidth(2)
axs.spines['left'].set_linewidth(2)
axs.set_xlabel("Set of combination")
axs.set_ylabel("Accuracy")
axs.set_xticks(labels, fontsize=20, color='#333')
axs.set_ylim(0.7,0.88)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{height}',
             ha='center', va='bottom')

# 获取当前的x轴刻度标签
ax = plt.gca()
xticks = ax.get_xticks()
xticklabels = ax.get_xticklabels()

# 定义一个颜色列表，默认为黑色
colors = ['black'] * len(hyper_acc)
# 将第12个标签设置为红色
colors[hyper_acc.index(max(hyper_acc))] = 'red'

# 应用颜色到刻度标签
for tick, color in zip(xticklabels, colors):
    tick.set_color(color)

    
def plt_hyper_comb(data):
    fig, axs = plt.subplots(1, 1, figsize=(8, 6))
    labels=range(0,len(hyper_acc),1)

    acc=[round(x, 2) for x in hyper_acc]

    # 子图1
    bars = axs.bar(labels, acc, alpha=0.6)
    axs.plot(labels, acc, label="Mean Reliability", marker="s")
    axs.set_title("Combination of hyperparameter",fontsize=14, fontweight='bold')
    axs.spines['top'].set_linewidth(2)
    axs.spines['right'].set_linewidth(2)
    axs.spines['bottom'].set_linewidth(2)
    axs.spines['left'].set_linewidth(2)
    axs.set_xlabel("Set of combination")
    axs.set_ylabel("Accuracy")
    axs.set_xticks(labels, fontsize=20, color='#333')
    axs.set_ylim(0.7,0.88)

    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{height}',
                 ha='center', va='bottom')

    # 获取当前的x轴刻度标签
    ax = plt.gca()
    xticks = ax.get_xticks()
    xticklabels = ax.get_xticklabels()

    # 定义一个颜色列表，默认为黑色
    colors = ['black'] * len(hyper_acc)
    # 将第12个标签设置为红色
    colors[data.index(max(data))] = 'red'

    # 应用颜色到刻度标签
    for tick, color in zip(xticklabels, colors):
        tick.set_color(color)
        
plt_hyper_comb(hyper_acc)
```

```python

s1_acc = [0.9489,0.9038,0.9107,0.8950,0.8780,0.8933,0.8957,0.9150,0.9378]
s2_acc = [0.88,0.8475,0.8717,0.8750,0.8720,0.8488,0.88,0.86,0.8550]
s3_acc = [0.9617,0.9250,0.9179,0.9008,0.8910,0.8908,0.9007,0.9225,0.9306]

labels= ['-4','-3','-2','-1','0','1','2','3','4']

plt.figure(figsize=(8, 5))

plt.plot(s1_acc, label='Oversampling', marker='o', linewidth=2, color='blue')
plt.plot(s2_acc, label='Undersampling', marker='o', linewidth=2, color='green')
plt.plot(s3_acc, label='SMOTE', marker='o', linewidth=2, color='red')

# 3. 添加标题和坐标轴标签
plt.title('Strategy Accuracy with different imbalance degree')
plt.xlabel('Imbalance degree')
plt.ylabel('Accuracy')
#plt.xticks([i for i in range(9)], label)

# 4. 添加网格、图例、以及适当的坐标轴范围（可选）
plt.grid(True)
plt.legend(loc='best')
plt.xticks(np.arange(9), labels)
plt.show()

def plt_acc_imbalance(data):
    labels= ['-4','-3','-2','-1','0','1','2','3','4']

    plt.figure(figsize=(8, 5))

    plt.plot(data[0], label='Oversampling', marker='o', linewidth=2, color='blue')
    plt.plot(data[1], label='Undersampling', marker='o', linewidth=2, color='green')
    plt.plot(data[2], label='SMOTE', marker='o', linewidth=2, color='red')

    # 3. 添加标题和坐标轴标签
    plt.title('Strategy Accuracy with different imbalance degree')
    plt.xlabel('Imbalance degree')
    plt.ylabel('Accuracy')
    #plt.xticks([i for i in range(9)], label)

    # 4. 添加网格、图例、以及适当的坐标轴范围（可选）
    plt.grid(True)
    plt.legend(loc='best')
    plt.xticks(np.arange(9), labels)
    plt.show()

accs=[s1_acc,s2_acc,s3_acc]
plt_acc_imbalance(accs)
```

```python
s_avg_acc = [np.array(s1_acc).mean(),np.array(s2_acc).mean(),np.array(s3_acc).mean()]

fig, axs = plt.subplots(1, 1, figsize=(6, 4))
labels=['Oversampling','Undersampling','SMOTE']

s_avg_acc=[round(x, 2) for x in s_avg_acc]


# 子图1
bars = axs.bar(labels, s_avg_acc, alpha=0.6, color="green")
#axs.plot(labels, s_avg_acc, label="Mean Reliability", color="green", marker="s")
axs.set_title("Average accuracy of strategy",fontsize=14, fontweight='bold')
axs.spines['top'].set_linewidth(1)
axs.spines['right'].set_linewidth(1)
axs.spines['bottom'].set_linewidth(2)
axs.spines['left'].set_linewidth(2)
axs.set_ylabel("Accuracy")
axs.set_xticks(labels, fontsize=20, color='#333')
axs.set_ylim(0.8,1)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{height}',
             ha='center', va='bottom')

# 获取当前的x轴刻度标签
ax = plt.gca()
xticks = ax.get_xticks()
xticklabels = ax.get_xticklabels()

# 定义一个颜色列表，默认为黑色
colors = ['black'] * len(hyper_acc)
# 将第12个标签设置为红色
colors[11] = 'red'

# 应用颜色到刻度标签
for tick, color in zip(xticklabels, colors):
    tick.set_color(color)
    
def plt_s_avg(data):
    fig, axs = plt.subplots(1, 1, figsize=(6, 4))
    labels=['Oversampling','Undersampling','SMOTE']

    value=[round(x, 2) for x in data]


    # 子图1
    bars = axs.bar(labels, value, alpha=0.6, color="green")
    #axs.plot(labels, s_avg_acc, label="Mean Reliability", color="green", marker="s")
    axs.set_title("Average accuracy of strategy",fontsize=14, fontweight='bold')
    axs.spines['top'].set_linewidth(1)
    axs.spines['right'].set_linewidth(1)
    axs.spines['bottom'].set_linewidth(2)
    axs.spines['left'].set_linewidth(2)
    axs.set_ylabel("Accuracy")
    axs.set_xticks(labels, fontsize=20, color='#333')
    axs.set_ylim(0.8,1)

    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{height}',
                 ha='center', va='bottom')


plt_s_avg(s_avg_acc)       
```
